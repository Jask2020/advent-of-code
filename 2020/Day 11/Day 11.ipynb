```python
state = [frcopy]
i = 0
rows = len(frcopy)
seats = len(frcopy[0])
while True:
    newstate = [list('x'*seats) for z in range(rows)]
    for j in range(rows):
        for k in range(seats):
            adjs = see(state[i], k, j, seats, rows)
            if state[i][j][k] == 'L':
                newstate[j][k] = 'L'
                if '#' not in adjs:
                    newstate[j][k] = '#'
            if state[i][j][k] == '#':
                newstate[j][k] = '#'
                if adjs.count('#') >= 5:
                    newstate[j][k] = 'L'
            if state[i][j][k] == '.':
                newstate[j][k] = '.'
            #if j==9 and k==5:
                #print(adjs)
    state.append(newstate)
    i += 1
    if state[i] in state[:-2]:
        print('error')
        for row in state[i]:
            print("".join(row))
        break
    if state[i] == state[i-1]:
        for row in state[i]:
            print("".join(row))
        print("\n")

        break
               
            
count = 0
for row in state[-1]:
    for seat in row:
        if seat == '#':
            count += 1
            
print(count)

```

```python
def see(curr, k, j, seats, rows):
    adjs = ['.','.','.','.','.','.','.','.']
    for x in range(k-1,-1,-1):
        if curr[j][x] != '.':
            adjs[0]=curr[j][x]
            break
    for x in range(k+1, seats):
        if curr[j][x] != '.':
            adjs[1]=curr[j][x]
            break
    for x in range(j-1,-1,-1):
        if curr[x][k] != '.':
            adjs[2]=curr[x][k]
            break
    for x in range(j+1, rows):
        if curr[x][k] != '.':
            adjs[3] = curr[x][k]
            break
    x,y=1,1
    while (x <= j and y <= k):
        if curr[j-x][k-y] != '.':
            adjs[4]=curr[j-x][k-y]
            break
        x += 1
        y += 1
    x,y=1,1
    while (x < rows-j and y <= k):
        if curr[j+x][k-y] != '.':
            adjs[5]=curr[j+x][k-y]
            break
        x += 1
        y += 1
    x,y=1,1
    while (x <= j and y < seats-k):
        if curr[j-x][k+y] != '.':
            adjs[6]=curr[j-x][k+y]
            break
        x += 1
        y += 1
    x,y=1,1
    while (x < rows-j and y < seats-k):
        if curr[j+x][k+y] != '.':
            adjs[7]=curr[j+x][k+y]
            break
        x += 1
        y += 1
    return adjs
```
