```python
dirc = 1
pos = [0,0]

for move in fr:
    if move[0] == 'N':
        pos[1] += int(move[1:])
    if move[0] == 'S':
        pos[1] -= int(move[1:])
    if move[0] == 'E':
        pos[0] += int(move[1:])
    if move[0] == 'W':
        pos[0] -= int(move[1:])
    if move[0] == 'R':
        dirc = (dirc+(int(move[1:])/90))%4
    if move[0] == 'L':
        dirc = (dirc-(int(move[1:])/90))%4
    if move[0] == 'F':
        if dirc==1:
            pos[0] += int(move[1:])
        if dirc==3:
            pos[0] -= int(move[1:])
        if dirc==0:
            pos[1] += int(move[1:])
        if dirc==2:
            pos[1] -= int(move[1:])
            
print(abs(pos[0]), abs(pos[1]))
    
```
