```python
f = open("doainput6.txt")
```

```python
fr = f.read()

lst = fr.split("\n\n")

lst
```

```python
#Part 1
count = 0

for group in lst:
    dicte = {}
    for x in group:
        if x in dicte.keys():
            continue
        else:
            dicte[x] = 1
    if '\n' in dicte.keys():
        count += len(dicte.keys())-1
    else:
        count += len(dicte.keys())
        
print(count)
            
```

```python
#Part 2
count2 = 0

for group in lst:
    people = group.split()
    dicte = {}
    for x in group:
        if x in dicte.keys():
            dicte[x] += 1
        else:
            dicte[x] = 1  
    for ltr in dicte.keys():
        if dicte[ltr] == len(people):
            count2 += 1

print(count2-1) # to account for whitespace
```
