```python
f = open("input8.txt")

fr = f.read().split('\n')

fr
```

```python
visited = []
idx = 0 
acc = 0
while idx < len(fr):
    print(fr[idx],idx)
    if idx in visited:
        break
    else:
        visited.append(idx)
    if fr[idx][:3] == 'acc':
        acc += int(fr[idx][4:])
    
    if fr[idx][:3] == 'jmp':
        idx += int(fr[idx][4:])
        continue
    idx+=1 

print(acc)
        
```
